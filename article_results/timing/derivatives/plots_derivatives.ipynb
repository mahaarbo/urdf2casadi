```python
output_file("dodged_bars.html")

dynamics = ['G', 'C', 'ID', 'M', 'FD (CRBA)', 'FD (ABA)']
#years = ['pybullet', 'RBDL', 'u2c']

data = {'dynamics' : dynamics,
        'u2c'   : [ur5_g, ur5_c, ur5_id, ur5_m, ur5_fdcrba, ur5_fdaba],
        'u2c derivatives'   : [ur5_g_dq, ur5_c_dq, ur5_id_dq, ur5_m_dq, ur5_fdcrba_dq, ur5_fdaba_dq]}

source = ColumnDataSource(data=data)

ur5_dq_hist = figure(x_range=dynamics, y_range=(0, 50), plot_height=260, title="6-dof UR5",
           toolbar_location=None, tools="", plot_width = 660)

ur5_dq_hist.vbar(x=dodge('dynamics', -0.10, range=ur5_dq_hist.x_range), top='u2c', width=0.2, source=source,
       color=palette2[2],  legend=value("u2c"))

ur5_dq_hist.vbar(x=dodge('dynamics',  +0.10,  range=ur5_dq_hist.x_range), top='u2c derivatives', width=0.2, source=source,
       color=palette2[0],  legend=value("u2c derivatives"))


labels = LabelSet(x='dynamics', level='glyph',
        x_offset=-13.5, y_offset=0, source=source, render_mode='canvas')

ur5_dq_hist.x_range.range_padding = 0.
ur5_dq_hist.xgrid.grid_line_color = None
ur5_dq_hist.legend.location = "top_left"
ur5_dq_hist.legend.orientation = "horizontal"
ur5_dq_hist.yaxis.axis_label = 'median evaluation time [us]'
ur5_dq_hist.axis.minor_tick_line_color = None

export_png(ur5_dq_hist, filename="plots/ur5_dq_hist_50.png")
```
