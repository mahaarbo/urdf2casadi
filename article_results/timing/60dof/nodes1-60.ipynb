```python
import numpy as np 
import urdf2casadi.urdfparser as u2c
from urdf_parser_py.urdf import URDF, Pose

ndofs = 60
urdf_nr = list(range(ndofs + 1))
urdf_nr.pop(0)

#storage for timing data
c_nodes = [None]*ndofs
g_nodes = [None]*ndofs
id_nodes = [None]*ndofs
m_nodes = [None]*ndofs
fdaba_nodes = [None]*ndofs
fdcrba_nodes = [None]*ndofs

nitr = 100
nrepeat = 100
    
for i in range(ndofs):    
    print i
    path_to_urdf = "/home/lillmaria/urdf2casadi/examples/timing/urdf4timing/" + str(urdf_nr[i]) + "dof.urdf"
    end_link = "link" + str(urdf_nr[i])
    root = 'base_link'
    tip = end_link
    gravity_u2c = [0, 0, -9.81]

    asd = u2c.URDFparser()
    robot_desc = asd.from_file(path_to_urdf)
    
    jlist, names, qmax, qmin = asd.get_joint_info(root, tip)
    njoints = len(jlist)
    
    g = asd.get_gravity_rnea(root, tip, gravity_u2c)
    g_nodes[i] = g.n_nodes()
    
    c = asd.get_coriolis_rnea(root, tip)
    c_nodes[i] = c.n_nodes()
    
    m = asd.get_inertia_matrix_crba(root, tip)
    m_nodes[i] = m.n_nodes()
    
    fd_aba = asd.get_forward_dynamics_aba(root, tip, gravity_u2c)
    fdaba_nodes[i] = fd_aba.n_nodes()
    
    fd_crba = asd.get_forward_dynamics_crba(root, tip, gravity_u2c)
    fdcrba_nodes[i] = fd_crba.n_nodes()
    
    iid = asd.get_inverse_dynamics_rnea(root, tip, gravity_u2c)
    id_nodes[i] = iid.n_nodes()
    
    
    
print "g_nodes = ", g_nodes
print "c_nodes = ", c_nodes
print "m_nodes = ", m_nodes
print "id_nodes = ", id_nodes
print "fdaba_nodes = ", fdaba_nodes
print "fdcrba_nodes = ", fdcrba_nodes







```
